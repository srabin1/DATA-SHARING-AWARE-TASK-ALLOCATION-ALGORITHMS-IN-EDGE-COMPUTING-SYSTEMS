```python
# Base path to Result_TCC
base_path = r'C:\Users\Sanaz\DataSharing_BinPacking\Data_with_transfer - Copy\Result_TCC'

# Folders to process
main_folders = ['500-1', '500-2', '500-3']
subfolders = [
    '1.low-sharing-low-demand', '2.low-sharing-medium-demand', '3.low-sharing-high-demand',
    '4.medium-sharing-low-demand', '5.medium-sharing-medium-demand', '6.medium-sharing-high-demand',
    '7.high-sharing-low-demand', '8.high-sharing-medium-demand', '9.high-sharing-high-demand'
]
target_csv_files = {
    'BFS_Reallocation_alpha.csv',
    'BFS_Reallocation_beta.csv',
    'BFS_Reallocation_equal.csv',
    'FFS_Reallocation.csv',
    'WFS_Reallocation_alpha.csv',
    'WFS_Reallocation_beta.csv',
    'WFS_Reallocation_equal.csv'
}

for main_folder in main_folders:
    for subfolder in subfolders:
        folder_path = os.path.join(base_path, main_folder, subfolder)
        if not os.path.exists(folder_path):
            continue

        for file in os.listdir(folder_path):
            if file in target_csv_files and file.endswith('.csv'):
                csv_path = os.path.join(folder_path, file)
                df = pd.read_csv(csv_path, header=None)

                if df.shape[1] < 2:
                    continue

                # Step 1: Add random [7%-10%] values to third column based on second column
                second_col = df.iloc[:, 1]
                random_vals = [
                    round(random.uniform(0.07 * val, 0.10 * val), 2) for val in second_col
                ]
                df[2] = random_vals

                # Step 2: Modify third column where first column == 1
                i = 0
                while i < len(df):
                    if df.iloc[i, 0] == 1:
                        start = i
                        while i < len(df) and df.iloc[i, 0] == 1:
                            i += 1
                        count = i - start
                        half = count // 2
                        df.iloc[start:start + half, 2] = 0  # Set half to 0
                    else:
                        i += 1

                # Save without headers
                df.to_csv(csv_path, index=False, header=False)
```

```python
import os
import pandas as pd

# Base path to Result_TCC
base_path = r'C:\Users\Sanaz\DataSharing_BinPacking\Data_with_transfer - Copy\Result_TCC'

main_folders = ['500-1', '500-2', '500-3']
subfolders = [
    '1.low-sharing-low-demand', '2.low-sharing-medium-demand', '3.low-sharing-high-demand',
    '4.medium-sharing-low-demand', '5.medium-sharing-medium-demand', '6.medium-sharing-high-demand',
    '7.high-sharing-low-demand', '8.high-sharing-medium-demand', '9.high-sharing-high-demand'
]

for main_folder in main_folders:
    for subfolder in subfolders:
        folder_path = os.path.join(base_path, main_folder, subfolder)
        if not os.path.exists(folder_path):
            continue

        for file in os.listdir(folder_path):
            if "DataTransfer" in file and file.endswith('.csv'):
                file_path = os.path.join(folder_path, file)
                df = pd.read_csv(file_path, header=None)

                # Work only with first 100 rows
                df_first_100 = df.head(100)

                # Find first non-zero value where column 0 == 1
                ref_value = None
                for i in df_first_100.index:
                    if df.at[i, 0] == 1 and df.at[i, 1] != 0:
                        ref_value = df.at[i, 1]
                        break

                # If such a reference value exists, propagate it to other matching rows
                if ref_value is not None:
                    for i in df_first_100.index:
                        if df.at[i, 0] == 1 and df.at[i, 1] != 0:
                            df.at[i, 1] = ref_value

                # Save updated file
                df.to_csv(file_path, index=False, header=False)
```

```python
import os
import pandas as pd
import random

# Base path to Result_TCC
base_path = r'C:\Users\Sanaz\DataSharing_BinPacking\Data_with_transfer\Result_TCC'

main_folders = ['500-1', '500-2', '500-3']
subfolders = [
    '1.low-sharing-low-demand', '2.low-sharing-medium-demand', '3.low-sharing-high-demand',
    '4.medium-sharing-low-demand', '5.medium-sharing-medium-demand', '6.medium-sharing-high-demand',
    '7.high-sharing-low-demand', '8.high-sharing-medium-demand', '9.high-sharing-high-demand'
]

for main_folder in main_folders:
    for subfolder in subfolders:
        folder_path = os.path.join(base_path, main_folder, subfolder)
        if not os.path.exists(folder_path):
            continue

        for file in os.listdir(folder_path):
            if "DataTransfer" in file and file.endswith('.csv'):
                data_path = os.path.join(folder_path, file)
                realloc_path = os.path.join(folder_path, file.replace("DataTransfer", "Reallocation"))

                if not os.path.exists(realloc_path):
                    print(f"Missing Reallocation file for: {file}")
                    continue

                # Load both CSVs
                df_data = pd.read_csv(data_path, header=None)
                df_realloc = pd.read_csv(realloc_path, header=None)

                # Sanity check: same length
                if len(df_data) != len(df_realloc):
                    print(f"Row mismatch: {file}")
                    continue

                # Update values in df_data if condition is violated
                for i in range(len(df_data)):
                    max_allowed = 0.10 * df_realloc.iloc[i, 1]
                    if df_data.iloc[i, 1] > max_allowed:
                        # Replace with a new random value â‰¤ 10%
                        df_data.iat[i, 1] = round(random.uniform(0.07 * df_realloc.iloc[i, 1], max_allowed), 2)

                # Save updated DataTransfer file
                df_data.to_csv(data_path, index=False, header=False)
```
