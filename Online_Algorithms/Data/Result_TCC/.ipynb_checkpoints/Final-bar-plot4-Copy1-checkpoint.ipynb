```python
import os
import pandas as pd
import numpy as np

main_folder_path = 'C:/Users/sanaz/DataSharing_BinPacking/Data/Result_TCC'
subfolder_names = ['500-1', '500-2', '500-3']
subsubfolder_names = [
    '1.low-sharing-low-demand',
    '2.low-sharing-medium-demand',
    '3.low-sharing-high-demand',
    '4.medium-sharing-low-demand',
    '5.medium-sharing-medium-demand',
    '6.medium-sharing-high-demand',
    '7.high-sharing-low-demand',
    '8.high-sharing-medium-demand',
    '9.high-sharing-high-demand'
]

csv_files_1 = ['FFS-Online-Sharing.csv', 'BFS-Online-Sharing-equal.csv', 'WFS-Online-Sharing-equal.csv']
csv_files_2 = ['FF-Online-Sharing-Oblivious.csv', 'BF-Online-Sharing-Oblivious-equal.csv', 'WF-Online-Sharing-Oblivious-equal.csv']

# Initialize variables to store max values
x_max, y_max, z_max = None, None, None
x_prime_max, y_prime_max, z_prime_max = None, None, None

# Initialize variables to store calculated values
calculation_1, calculation_2, calculation_3 = None, None, None

# Initialize variable to store average value
average_value = None

# Initialize variables to store squared differences for standard deviation
squared_diff_1, squared_diff_2, squared_diff_3 = 0, 0, 0

# Initialize count of non-None values
count_non_none = 0

for subfolder_name in subfolder_names:
    subfolder_path = os.path.join(main_folder_path, subfolder_name)
    
    if os.path.exists(subfolder_path):
        print(f"Entering subfolder: {subfolder_name}")
        
        for subsubfolder_name in subsubfolder_names:
            full_subsubfolder_path = os.path.join(subfolder_path, subsubfolder_name)
            
            if os.path.exists(full_subsubfolder_path):
                print(f"Entering sub-subfolder: {subsubfolder_name}")
                
                for csv_file in csv_files_1:
                    csv_file_path = os.path.join(full_subsubfolder_path, csv_file)
                    
                    if os.path.exists(csv_file_path):
                        print(f"Processing CSV file: {csv_file}")
                        
                        # Read the CSV file using pandas
                        df = pd.read_csv(csv_file_path)
                        
                        # Get the maximum value in the first column
                        max_value = df.iloc[:, 0].max()
                        
                        # Assign the max value to the corresponding variable
                        if csv_file == 'FFS-Online-Sharing.csv':
                            x_max = max_value
                        elif csv_file == 'BFS-Online-Sharing-equal.csv':
                            y_max = max_value
                        elif csv_file == 'WFS-Online-Sharing-equal.csv':
                            z_max = max_value
                        
                        # Or perform any other actions you need with the data in the CSV file
                        
                    else:
                        print(f"CSV file {csv_file} does not exist.")
                
                for csv_file_prime in csv_files_2:
                    csv_file_prime_path = os.path.join(full_subsubfolder_path, csv_file_prime)
                    
                    if os.path.exists(csv_file_prime_path):
                        print(f"Processing CSV file: {csv_file_prime}")
                        
                        # Read the CSV file using pandas
                        df_prime = pd.read_csv(csv_file_prime_path)
                        
                        # Get the maximum value in the first column
                        max_value_prime = df_prime.iloc[:, 0].max()
                        
                        # Assign the max value to the corresponding variable
                        if csv_file_prime == 'FF-Online-Sharing-Oblivious.csv':
                            x_prime_max = max_value_prime
                        elif csv_file_prime == 'BF-Online-Sharing-Oblivious-equal.csv':
                            y_prime_max = max_value_prime
                        elif csv_file_prime == 'WF-Online-Sharing-Oblivious-equal.csv':
                            z_prime_max = max_value_prime
                        
                        # Or perform any other actions you need with the data in the CSV file
                        
                    else:
                        print(f"CSV file {csv_file_prime} does not exist.")
                
                # Calculate the specified values
                if x_prime_max is not None and x_max is not None:
                    calculation_1 = (x_prime_max - x_max) / x_prime_max
                    
                if y_prime_max is not None and y_max is not None:
                    calculation_2 = (y_prime_max - y_max) / y_prime_max
                    
                if z_prime_max is not None and z_max is not None:
                    calculation_3 = (z_prime_max - z_max) / z_prime_max
                
                # Sum up the non-None values
                if calculation_1 is not None:
                    average_value = calculation_1
                    count_non_none += 1
                if calculation_2 is not None:
                    average_value = (average_value + calculation_2) if average_value is not None else calculation_2
                    count_non_none += 1
                if calculation_3 is not None:
                    average_value = (average_value + calculation_3) if average_value is not None else calculation_3
                    count_non_none += 1
                
                # Calculate squared differences for standard deviation
                squared_diff_1 += (calculation_1 - average_value)**2 if calculation_1 is not None else 0
                squared_diff_2 += (calculation_2 - average_value)**2 if calculation_2 is not None else 0
                squared_diff_3 += (calculation_3 - average_value)**2 if calculation_3 is not None else 0
                
                # Or perform any other actions you need with the calculated values
                
            else:
                print(f"Sub-subfolder {subsubfolder_name} does not exist.")
                
    else:
        print(f"Subfolder {subfolder_name} does not exist.")

# Calculate the standard deviation
if count_non_none > 1:
    standard_deviation = np.sqrt((squared_diff_1 + squared_diff_2 + squared_diff_3) / (count_non_none - 1))
else:
    standard_deviation = None

# Print or use the stored max values, calculated values, average value, and standard deviation as needed
print(f"Max value for FFS-Online-Sharing.csv (x): {x_max}")
print(f"Max value for BFS-Online-Sharing-equal.csv (y): {y_max}")
print(f"Max value for WFS-Online-Sharing-equal.csv (z): {z_max}")

print(f"Max value for FF-Online-Sharing-Oblivious.csv (x'): {x_prime_max}")
print(f"Max value for BF-Online-Sharing-Oblivious-equal.csv (y'): {y_prime_max}")
print(f"Max value for WF-Online-Sharing-Oblivious-equal.csv (z'): {z_prime_max}")

print(f"Calculation 1: (x'-x)/x': {calculation_1}")
print(f"Calculation 2: (y'-y)/y': {calculation_2}")
print(f"Calculation 3: (z'-z)/z': {calculation_3}")

print(f"Average of Calculations: {average_value}")
print(f"Standard Deviation of Calculations: {standard_deviation}")
```

```python
import os
import pandas as pd
import numpy as np

main_folder_path = 'C:/Users/sanaz/DataSharing_BinPacking/Data/Result_TCC'
subfolder_names = ['500-1', '500-2', '500-3']
subsubfolder_names = [
    '1.low-sharing-low-demand',
    '2.low-sharing-medium-demand',
    '3.low-sharing-high-demand',
    '4.medium-sharing-low-demand',
    '5.medium-sharing-medium-demand',
    '6.medium-sharing-high-demand',
    '7.high-sharing-low-demand',
    '8.high-sharing-medium-demand',
    '9.high-sharing-high-demand'
]

csv_files_1 = ['WFS-Online-Sharing-equal.csv']
csv_files_2 = ['WF-Online-Sharing-Oblivious-equal.csv']

# Initialize variables to store max values
x_max, y_max, z_max = None, None, None
x_prime_max, y_prime_max, z_prime_max = None, None, None

# Initialize variables to store calculated values
calculation_1, calculation_2, calculation_3 = None, None, None

# Initialize variables to store average and standard deviation values for each subsubfolder
average_values = {}
standard_deviation_values = {}

for subfolder_name in subfolder_names:
    subfolder_path = os.path.join(main_folder_path, subfolder_name)
    
    if os.path.exists(subfolder_path):
        print(f"Entering subfolder: {subfolder_name}")
        
        for subsubfolder_name in subsubfolder_names:
            full_subsubfolder_path = os.path.join(subfolder_path, subsubfolder_name)
            
            if os.path.exists(full_subsubfolder_path):
                print(f"Entering sub-subfolder: {subsubfolder_name}")
                
                # Initialize variables to store values for each subsubfolder
                subsubfolder_values = []
                
                for csv_file in csv_files_1:
                    csv_file_path = os.path.join(full_subsubfolder_path, csv_file)
                    
                    if os.path.exists(csv_file_path):
                        print(f"Processing CSV file: {csv_file}")
                        
                        # Read the CSV file using pandas
                        df = pd.read_csv(csv_file_path)
                        
                        # Get the maximum value in the first column
                        max_value = df.iloc[:, 0].max()
                        
                        # Assign the max value to the corresponding variable
                        if csv_file == 'WFS-Online-Sharing-equal.csv':
                            x_max = max_value
                        elif csv_file == 'WFS-Online-Sharing-equal.csv':
                            y_max = max_value
                        elif csv_file == 'WFS-Online-Sharing-equal.csv':
                            z_max = max_value
                        
                        # Or perform any other actions you need with the data in the CSV file
                        
                    else:
                        print(f"CSV file {csv_file} does not exist.")
                
                for csv_file_prime in csv_files_2:
                    csv_file_prime_path = os.path.join(full_subsubfolder_path, csv_file_prime)
                    
                    if os.path.exists(csv_file_prime_path):
                        print(f"Processing CSV file: {csv_file_prime}")
                        
                        # Read the CSV file using pandas
                        df_prime = pd.read_csv(csv_file_prime_path)
                        
                        # Get the maximum value in the first column
                        max_value_prime = df_prime.iloc[:, 0].max()
                        
                        # Assign the max value to the corresponding variable
                        if csv_file_prime == 'WF-Online-Sharing-Oblivious-equal.csv':
                            x_prime_max = max_value_prime
                        elif csv_file_prime == 'WF-Online-Sharing-Oblivious-equal.csv':
                            y_prime_max = max_value_prime
                        elif csv_file_prime == 'WF-Online-Sharing-Oblivious-equal.csv':
                            z_prime_max = max_value_prime
                        
                        # Or perform any other actions you need with the data in the CSV file
                        
                    else:
                        print(f"CSV file {csv_file_prime} does not exist.")
                
                # Calculate the specified values
                if x_prime_max is not None and x_max is not None:
                    calculation_1 = (x_prime_max - x_max) / x_prime_max
                    subsubfolder_values.append(calculation_1)
                    
                if y_prime_max is not None and y_max is not None:
                    calculation_2 = (y_prime_max - y_max) / y_prime_max
                    subsubfolder_values.append(calculation_2)
                    
                if z_prime_max is not None and z_max is not None:
                    calculation_3 = (z_prime_max - z_max) / z_prime_max
                    subsubfolder_values.append(calculation_3)
                
                # Store values for each subsubfolder
                average_values[subsubfolder_name] = np.mean(subsubfolder_values)
                standard_deviation_values[subsubfolder_name] = np.std(subsubfolder_values, ddof=1) if len(subsubfolder_values) > 1 else None
                
                # Or perform any other actions you need with the calculated values
                
            else:
                print(f"Sub-subfolder {subsubfolder_name} does not exist.")
                
    else:
        print(f"Subfolder {subfolder_name} does not exist.")

# Print or use the stored max values, calculated values, average, and standard deviation values for each subsubfolder as needed
for subsubfolder_name in subsubfolder_names:
    print(f"Sub-subfolder: {subsubfolder_name}")
    print(f"Average: {average_values.get(subsubfolder_name, 'N/A')}")
    print(f"Standard Deviation: {standard_deviation_values.get(subsubfolder_name, 'N/A')}")
    print("-----------------------")
```

```python
import os
import pandas as pd

# Define the main folder path and subfolder names
main_folder_path = 'C:/Users/sanaz/DataSharing_BinPacking/Data/Result_TCC'
subfolder_names = ['500-1', '500-2', '500-3']
subsubfolder_names = ['1.low-sharing-low-demand']

# Initialize an empty DataFrame to store the results
result_df = pd.DataFrame()

# Loop through the main folder, subfolders, and subsubfolders
for i, subfolder in enumerate(subfolder_names, 1):
    for subsubfolder in subsubfolder_names:
        # Construct the full path to the CSV file
        csv_file_path = os.path.join(main_folder_path, subfolder, subsubfolder, 'BFS-Online-Sharing-equal.csv')
        
        # Check if the file exists
        if os.path.exists(csv_file_path):
            # Read the CSV file
            df = pd.read_csv(csv_file_path)
            
            # Extract the first column
            first_column = df.iloc[:, 0]
            
            # Add a label to each column
            column_label = f'DSA-BF-equal-{i}'
            result_df = pd.concat([result_df, first_column.rename(column_label)], axis=1)

# Write the result DataFrame to a new CSV file
result_df.to_csv('BFS-equal compare.csv', index=False)

```
